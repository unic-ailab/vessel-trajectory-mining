# Interactive heat map for each week

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium 
from folium import plugins
from folium.plugins import HeatMap
from collections import Counter
from datetime import datetime as dt

#loading the original dataset
df_acc = pd.read_csv("nari_dynamic.csv",delimiter=',')

map_hooray = folium.Map(location=[48, -4.4],
                    zoom_start = 10) 

# conert the values into floats
df_acc['Latitude'] = df_acc['lat'].astype(float)
df_acc['Longitude'] = df_acc['lon'].astype(float)

heat_df = df_acc[['Latitude', 'Longitude']]
```

```python
#convert the unix time into days in the form on YearMonthDay
df_acc['date'] = df_acc['time'].map(lambda x: dt.fromtimestamp(x).strftime('%Y%m%d'))
```

```python
days=df_acc['date']
days_uniq=days.unique()
```

```python
len(days_uniq)
```

```python
df_acc=df_acc.sort_values(by='date')
print(df_acc.head())
print(df_acc.tail())
```

```python
days_uniq
```

```python
df_acc['week']=np.nan
df_acc
```

```python
#add the the number of the week at each record
test=df_acc.copy()

#week
j=1
# number of day
c=1

for i in days_uniq:
    if c>=8:
        c=1
        j+=1

    test.loc[test.date == i, 'week'] = j 
    c+=1 
```

```python
test
```

```python
# Create weight column, using date
heat_df['Weight'] = test['week']
heat_df['Weight'] = heat_df['Weight'].astype(float)
heat_df = heat_df.dropna(axis=0, subset=['Latitude','Longitude', 'Weight'])
```

```python
#fist week
x=int(heat_df.iloc[0,-1])
#last week
y=int(heat_df.iloc[-1,-1])
print(x,y)
```

```python
#create a list of lists for the weeks
heat_data = [[[row['Latitude'],row['Longitude']] for index, row in heat_df[heat_df['Weight'] == i].iterrows()] for i in range(x,y+1)]
```

```python
# heat map with layers divided into weeks
hm = plugins.HeatMapWithTime(heat_data,auto_play=True,max_opacity=0.8)
print('0')
hm.add_to(map_hooray)
print('1')
map_hooray.save("my_time_h_map_full_weeks.html")
print('2')
```
