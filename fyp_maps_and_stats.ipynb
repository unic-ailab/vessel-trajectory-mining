# Time heat map

```python
from datetime import datetime as dt

#loading the csv for the first week of the dataset
df_acc = pd.read_csv("vessel.csv",delimiter=',')

map_hooray = folium.Map(location=[48, -4.4],
                    zoom_start = 13) 

# converting the data into float
df_acc['Latitude'] = df_acc['lat'].astype(float)
df_acc['Longitude'] = df_acc['lon'].astype(float)

#list of list for the heat map
heat_df = df_acc[['Latitude', 'Longitude']]
```

```python
#from unix time we convert the time to normal time and date
df_acc['datetime'] = df_acc['time'].map(lambda x: dt.fromtimestamp(x).strftime('%d/%m/%Y %H:%M:%S'))
df_acc[['Date','datetime']]=df_acc.datetime.str.split(" ",expand=True,)

# Create weight column, using date 
heat_df['Weight'] = df_acc['Date'].str[:2]
heat_df['Weight'] = heat_df['Weight'].astype(float)
heat_df = heat_df.dropna(axis=0, subset=['Latitude','Longitude', 'Weight'])
```

```python
# List comprehension to make out list of lists
heat_data = [[[row['Latitude'],row['Longitude']] for index, row in heat_df[heat_df['Weight'] == i].iterrows()] for i in range(0,8)]
```

```python
# creating a heat map with layers
# each layer contains the points for each day of the weel
hm = plugins.HeatMapWithTime(heat_data,auto_play=True,max_opacity=0.8)

hm.add_to(map_hooray)

map_hooray.save("my_time_h_map10.html")
```

# Distance algorithm 
## distance coverd by each vessel during week 1

```python
import geopy.distance

dist=df.copy()

#empty dictionary for the vessel id as key and the distance as value
dist_d=dict()

#loop through all the vessels
for i in total_trips['id']:
    td=dist.copy()
    
    #filter the data in each iteration for each vessel
    td = td[td['id']==i] 
    td = td[['lat', 'lon']]
    td = td.dropna(axis=0, subset=['lat','lon'])
    
    j=1
    dist_d[i]=0
    
    #find the first 2 coordinates
    try:
        #prev coords
        coords_1=td.iloc[j-1]
        #print(coords_1)
        #current coords
        coords_2=td.iloc[j]
        #print(coords_2)
    except:
        continue
        
    sum_d=0
    while j <len(td):
        #sum up all the distance between two consecutive coortinates
        sum_d+=geopy.distance.distance(coords_1, coords_2).km 

        #continue to the next set of coortinates
        j+=1
        try:
            coords_1=td.iloc[j-1]
            coords_2=td.iloc[j]
        except:
            print("out of bounds")
    #store the values  
    dist_d[i]=sum_d


print(dist_d)
```

```python
print(len(dist_d))
dist_d
```

```python
#conversion of the dictionary to df
dist_km=pd.DataFrame.from_dict(dist_d, orient='index')
dist_km.columns = ['distance_km']
dist_km['id'] = dist_km.index
dist_km=dist_km[['id','distance_km']]
dist_km
```

```python
#points per day
one_day=86400
i=begining_time
days=[]

#find the total number of points from a week for each day
while i <=finish_time:
    points_day=points_week.copy()
    i+=one_day
    points_day = points_day[points_day['time']<=i]
    points_day = points_day[['time']]
    points_day = points_day.dropna(axis=0, subset=['time'])
    days.append(len(points_day))

print(days)

temp_days=days.copy()
i=1
while i< len(days):
    temp_days[i]=temp_days[i]-days[i-1]
    i+=1
    
print(temp_days)
```

```python
#points per hour
one_hour=86400/24
j=begining_time
hours=[]

#find the total number of points of a week for each hour
while j <=finish_time:
    points_hour=points_week.copy()
    j+=one_hour
    points_hour = points_hour[points_hour['time']<=j] # Reducing data size so it runs faster
    points_hour = points_hour[['time']]
    points_hour = points_hour.dropna(axis=0, subset=['time'])
    hours.append(len(points_hour))



temp_hours=hours.copy()
j=1
while j< len(hours):
    temp_hours[j]=temp_hours[j]-hours[j-1]
    j+=1
    
print(temp_hours)
```
